```python
import numpy as np
```

```python
def step_function(soma):
    if soma > 0:
        return 1
    return 0
```

```python
def treinar_perceptron(dados_x, dados_y, pesos, taxa_aprendizado, epocas):
    print(f"--- INÍCIO DO TREINAMENTO ---")
    print(f"Pesos Iniciais: {pesos}")
    print(f"Taxa de aprendizado (alpha): {taxa_aprendizado}\n")

    for epoca in range(1, epocas + 1):
        print(f"=== ÉPOCA {epoca} ===")
        erro_total = 0

        for i, x_entrada in enumerate(dados_x):
            x_com_bias = np.insert(x_entrada, 0, 1)
            u = np.dot(pesos, x_com_bias)
            y_obtido = step_function(u)

            y_esperado = dados_y[i]
            erro = y_esperado - y_obtido

            nomes = ["Joãozinho", "Huguinho", "Zezinho", "Luizinho"]
            print(f"Aluno: {nomes[i]} | Entrada (com bias): {x_com_bias}")
            print(f"   Soma: {u:.2f} -> Saída h(x): {y_obtido} | Esperado (y): {y_esperado} | Erro: {erro}")
            if erro != 0:
                delta = taxa_aprendizado * erro * x_com_bias
                pesos_antigos = pesos.copy()
                pesos = pesos + delta
                erro_total += abs(erro)
                print(f"   -> ATUALIZAÇÃO: w_antigo:{pesos_antigos} + (alpha*{erro}*x) = w_novo:{pesos}")
            else:
                print(f"   -> Sem atualização.")
            print("-" * 30)

        print(f"Fim do Ciclo {epoca}. Pesos finais da época: {pesos}\n")

    return pesos
```

```python
if __name__ == "__main__":
    X = np.array([
        [0, 0], # Joãozinho
        [0, 1], # Huguinho
        [1, 0], # Zezinho
        [1, 1]  # Luizinho
    ])
    y = np.array([0, 0, 1, 1])
    w_iniciais = np.array([0.0, 0.0, 0.0])
    taxa_aprendizado = 0.1
    ciclos = 2
    w_finais = treinar_perceptron(X, y, w_iniciais, taxa_aprendizado, ciclos)
    print("RESUMO FINAL DOS PESOS:")
    print(f"Bias (theta_0): {w_finais[0]}")
    print(f"Peso Estudou (theta_1): {w_finais[1]}")
    print(f"Peso Trabalho (theta_2): {w_finais[2]}")
```
